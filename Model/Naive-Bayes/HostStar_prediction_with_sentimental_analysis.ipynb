```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Reading the training data from local disk
Train_path='D:/Adarsh_Laptop/studetmater/5f828822-4--4-hotstar_dataset/train_data.json'
data = pd.read_json(Train_path,orient='values')
```

```python
data=data.T
data.head()
```

```python
#Lets define feature with only genres and titles columns for text classification 
X = data[['genres','titles']]
X.head()
```

```python
#Lets merge genres and titles columns in one columns and the column name will be 'text'
#X['text']= pd.concat(X.genres,X.titles)
new_X = X.copy()
new_X['text']= X.loc[:,'genres'].apply(lambda x : x)+ X.loc[:,'titles'].apply(lambda x:x)
```

```python
new_X.head()
```

```python
#Feature text columns will be 'text' so lets take it out
X_data = new_X.text
X_data.head()
```

```python
#Lets keep numric values but replace : and , with whitespace
X_data = X_data.apply(lambda x : x.replace(':',' ')).apply(lambda x:x.replace(',',' '))
X_data.head()
```

```python
#Lets prepare target data i.e y
y_data = data.segment
y_data.head()
```

```python
#lets encode these target values 
lr = LabelEncoder()
y_data=lr.fit_transform(y_data)
y_data[:50]
```

```python
#split the data into train and test sets
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_data,y_data,test_size=0.4,random_state=42)
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
#Lets vectorize the feature
vect= CountVectorizer()
vect.fit(X_train)
```

```python
X_train_dtm = vect.transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
vect.vocabulary_
```

```python
# MultinomialNB naive bayes 
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(alpha=1.0)
nb.fit(X_train_dtm,y_train)
```

```python
nb_pred = nb.predict(X_test_dtm)
```

```python
#print first 25 true and predicted values 
print 'Predicted',nb_pred[:25]
print 'True     ',y_test[:25]
```

```python
#lets see the accuracy score and other scores
from sklearn import metrics
nb_accuracy = metrics.accuracy_score(y_test,nb_pred)
nb_accuracy
```

```python
#Lets see the confusion matrix
pd.crosstab(y_test,nb_pred,rownames=['True'],colnames=['Predicted'],margins=True)
```
